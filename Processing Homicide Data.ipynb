```python
data=pd.read_csv(r"Medical_Examiner_Case_Archive%2C_2014_to_present.csv")
data.head()
```

```python
y=(data["INCIDENT_DATE"])
Year=[]
for i in y:
    i=str(i)
    i=i[0:4]
    i=float(i)
    Year.append(i)
print(Year)
```

```python
y=(data["MANNER"])
Manner=[]
for i in y:
    Manner.append(i)
print(Manner)
```

```python
y=(data["INCIDENT_ZIP"])
Zip=[]
for i in y:
    Zip.append(i)
print(Zip)
```
