```python
import pandas as pd
import os
import matplotlib as plt
import numpy as np
```

```python
data=pd.read_csv(r"Medical_Examiner_Case_Archive%2C_2014_to_present.csv")
data.head()
```

```python
# converting data on the incident date into a list
y=(data["INCIDENT_DATE"])
Year=[]
for i in y:
    i=str(i)
    i=i[0:4]
    i=float(i)
    Year.append(i)
```

```python
# converting data on the manner of death into a list
y=(data["MANNER"])
Manner=[]
for i in y:
    Manner.append(i)
```

```python
# converting data on the incident zip into a list
y=(data["INCIDENT_ZIP"])
Zip=[]
for i in y:
    Zip.append(i)
```

```python
# creating a list of the positions of all the deaths in each list that took place in 2020
twentytwenty=[]
for i in range(0,56979):
    if Year[i]==2020:
        twentytwenty.append(i)  
```

```python
# creating a list of the zip code where each 2020 homicide in cook county took place
Zips=[]
for i in twentytwenty:
    if (Manner[i])=="HOMICIDE":
        Zips.append(Zip[i])

        
        
```

```python
nZips=np.array(Zips)
L=np.unique(nZips)
```

```python
# printing how many homicides occured in each zip code of cook county in 2020
for p in L:
    print(p)
    print(Zips.count(p))
```
